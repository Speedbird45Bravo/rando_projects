```python
import datetime as dt
import pandas as pd
import requests
```

```python
# Classes for arrived and departed flights that return dataframes.
```

```python
class Arrivals:

  def __init__(self, arrival_airport, how_many):

    self.arrival_airport = arrival_airport
    self.how_many = how_many

    self.payload = {"airport":self.arrival_airport, "howMany":how_many}

    self.url = "http://flightxml.flightaware.com/json/FlightXML2/"
    self.api_key = "##### key #####"
    self.username = "##### username #####"

    self.js = requests.get(self.url + "Arrived", params=self.payload, auth=(self.username, self.api_key)).json()

    self.df = pd.DataFrame(self.js['ArrivedResult']['arrivals'])

    self.df['actualdeparturetime'] = self.df['actualdeparturetime'].apply(lambda x: dt.datetime.fromtimestamp(x))
    self.df['actualarrivaltime'] = self.df['actualarrivaltime'].apply(lambda x: dt.datetime.fromtimestamp(x))

  def display(self):
    return self.df
```

```python
class Departures:

  def __init__(self, departure_airport, how_many):

    self.departure_airport = departure_airport
    self.how_many = how_many

    self.payload = {"airport":self.departure_airport, "howMany":self.how_many}

    self.url = "http://flightxml.flightaware.com/json/FlightXML2/"
    self.api_key = "##### key #####"
    self.username = "##### username #####"

    self.js = requests.get(self.url + "Departed", params=self.payload, auth=(self.username, self.api_key)).json()

    self.df = pd.DataFrame(self.js['DepartedResult']['departures'])

    self.df['actualdeparturetime'] = self.df['actualdeparturetime'].apply(lambda x: dt.datetime.fromtimestamp(x))
    self.df['estimatedarrivaltime'] = self.df['estimatedarrivaltime'].apply(lambda x: dt.datetime.fromtimestamp(x))

  def display(self):
    return self.df
```

```python
Arrivals("KJFK",10).display()
```

```python
Departures("KJFK",10).display()
```
